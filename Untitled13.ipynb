```python
# Import essential libraries for data analysis and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Enable inline plotting and set visualization style
%matplotlib inline
sns.set(style="whitegrid")
```

```python
# Load Bremen weather data from CSV file
df = pd.read_csv('/content/bremen.csv')
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Convert 'datetime' column to datetime format
df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')

# Check for rows where datetime conversion failed
df[df['datetime'].isnull()]
```

```python
# Set 'datetime' column as the index of the dataframe
df.set_index('datetime', inplace=True)

# Sort the index to ensure chronological order
df.sort_index(inplace=True)
```

```python
# Display dataset information including column types and non-null counts
df.info()

# Check for missing values in each column
df.isnull().sum()
```

```python
# Convert selected columns to numeric, coercing errors
numeric_columns = ['tempmax', 'tempmin', 'temp', 'feelslikemax', 'feelslikemin', 'feelslike',
                   'dew', 'humidity', 'precip', 'snow', 'snowdepth', 'windgust', 'windspeed',
                   'winddir', 'sealevelpressure', 'cloudcover', 'visibility', 'solarradiation',
                   'solarenergy', 'uvindex', 'moonphase']

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Plot daily average temperature in Bremen
plt.figure(figsize=(12,6))
plt.plot(df['temp'], color='orange')
plt.title('Daily Average Temperature in Bremen')
plt.xlabel('Date')
plt.ylabel('Temperature (°F)')
plt.grid(True)
plt.show()
```

```python
# Calculate total monthly precipitation
monthly_precip = df['precip'].resample('M').sum()

# Plot monthly precipitation as a bar chart
plt.figure(figsize=(12,6))
monthly_precip.plot(kind='bar', color='blue')
plt.title('Monthly Precipitation in Bremen')
plt.xlabel('Month')
plt.ylabel('Precipitation (inches)')
plt.grid(True)
plt.show()
```

```python
# Display correlation heatmap between selected weather variables
selected_cols = ['tempmax', 'tempmin', 'temp', 'humidity', 'dew', 'windspeed', 'sealevelpressure', 'cloudcover']
plt.figure(figsize=(10,8))
sns.heatmap(df[selected_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Weather Variables')
plt.show()
```

```python
# Map months to seasons
df['month'] = df.index.month
df['season'] = df['month'].map({12:'Winter',1:'Winter',2:'Winter',
                                 3:'Spring',4:'Spring',5:'Spring',
                                 6:'Summer',7:'Summer',8:'Summer',
                                 9:'Autumn',10:'Autumn',11:'Autumn'})

# Plot temperature distribution by season
plt.figure(figsize=(10,6))
sns.boxplot(x='season', y='temp', data=df, palette='Set2')
plt.title('Temperature Distribution by Season')
plt.ylabel('Temperature (°F)')
plt.show()
```

## 📌 Key Findings

- **Temperature Trends:** Daily temperatures in Bremen during October 2025 ranged from lows of ~34°F to highs of ~64°F. The average temperature showed a slight warming trend mid-month.
- **Rainfall Patterns:** Rainfall occurred frequently, especially between October 4–17, with some days reaching over 0.1 inches. Rain probability was consistently high (100%) on rainy days.
- **Humidity Levels:** Humidity remained high throughout the month, often exceeding 85%, indicating a moist climate.
- **Seasonal Transition:** The data reflects a typical autumn transition in Bremen, with increasing cloud cover and reduced solar radiation toward the end of the month.
- **Correlations:** Temperature and dew point showed strong positive correlation, while humidity and temperature had a negative relationship.

These insights can support climate awareness, urban planning, and seasonal forecasting in northern Germany.
